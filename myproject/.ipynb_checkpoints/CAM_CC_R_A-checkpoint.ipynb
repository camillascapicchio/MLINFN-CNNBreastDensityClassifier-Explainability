```python
from vis.utils import utils
from keras import activations
from keras import models
from keras.models import load_model, Model
from vis.utils import utils
from matplotlib import pyplot as plt
from keras.preprocessing import image
from vis.visualization import visualize_saliency, overlay, visualize_activation, visualize_cam
import numpy as np
import scipy.misc
import matplotlib
import math
from scipy.ndimage import convolve
import os
import cv2

import warnings
warnings.filterwarnings('ignore')
```

```python
path='/media/camilla/DATA/jup/.venv/bin/CARTELLA_TESI/RESNET17/CC_R/'
model=load_model(path + '1/weights-improvement-46-0.80.h5')

model.summary()
```

```python
layer_idx = utils.find_layer_idx(model, 'dense_1')
model.layers[layer_idx].activation = activations.linear
modelnew = utils.apply_modifications(model)
penultimate_layer = utils.find_layer_idx(modelnew, 'add_12')
```

```python
classe = 'A'
path_test='/TestSet/CC_R/test/'
patient_ID = [d for d in os.listdir(path_test + classe) if os.path.isfile(os.path.join(path_test + classe, d))]
print(patient_ID)

for i in patient_ID: 
    print(i)
    indice=i.index('.')
    
    img = utils.load_img(path_test + classe +'/' + i, target_size=(650, 650))
    tensor=[img, img, img]
    img_reshape=np.stack(tensor, axis=2)
    mean = img_reshape.mean()
    #std = img.std()
    img_reshape = (img_reshape - mean)
    img_reshape = (img_reshape)*1./255
    
    img_resized = img_reshape[np.newaxis, ...]
    #img_resized = img_resized[..., np.newaxis]
    pred_class = np.argmax(model.predict(img_resized))
    print(pred_class)
    print(img_reshape.shape)
    #img_reshape = cv2.cvtColor(img_reshape, cv2.COLOR_RGB2RGBA)
    #plt.imsave(str(pred_class) + '_'+ i[:indice]+'_450.png', img_reshape, cmap=plt.cm.gray)
    scipy.misc.imsave(str(pred_class) + '_'+ i[:indice]+'_650.png', img_reshape)
    
    mappa=visualize_cam(modelnew, layer_idx, filter_indices=[pred_class], seed_input=img_resized, penultimate_layer_idx=penultimate_layer)

    plt.imsave(str(pred_class) + '_' + i[:indice] + '_map.png' , mappa, cmap='seismic')


```

```python
images = [d for d in os.listdir(path + '/mappe/' + classe) if os.path.isfile(os.path.join(path + '/mappe/' + classe, d))]

for i in images:
    indice=i.index('.')-4
    if i[indice:] == '_650.png':
        print(i)
        immagine=utils.load_img(i)
        plt.imshow(immagine)
        plt.show()
        print(immagine.shape)
        immagine = cv2.cvtColor(immagine, cv2.COLOR_RGB2RGBA)
        print(immagine.shape)
        #arrays=[immagine, immagine, immagine, immagine]
        #img_reshape=np.stack(arrays, axis=2)
        print(i[:indice] + '_map.png')
        mappa_color=utils.load_img(i[:indice] + '_map.png')
        print(mappa_color.shape)
        
        #mappa_overlay=overlay(img_reshape, mappa_color)
        mappa_overlay=cv2.addWeighted(immagine, 1, mappa_color, 1, 0)
        plt.imshow(mappa_overlay)
        plt.show()
        scipy.misc.imsave(i[:indice]+'_overlay.png', mappa_overlay)
        
        
      
    
```

```python
patient_ID = [d for d in os.listdir(path_test + classe) if os.path.isfile(os.path.join(path_test + classe, d))]

for i in patient_ID: 
    print(i)
    
    img = utils.load_img(path_test + classe +'/' + i, target_size=(650, 650))
    tensor=[img, img, img]
    img_reshape=np.stack(tensor, axis=2)
    mean = img_reshape.mean()
    #std = img.std()
    img_reshape = (img_reshape - mean)
    img_reshape = (img_reshape)*1./255
    img_resized = img_reshape[np.newaxis, ...]
    #img_resized = img_resized[..., np.newaxis]
    
    pred_class = np.argmax(model.predict(img_resized))
    print(pred_class)
    
    
    weights=model.layers[-1].get_weights()
    w=weights[0]
    w=w[:, pred_class]

    layer_output = model.layers[-5].output
    
    newModel = Model(model.inputs, layer_output)
    outputs=newModel.predict(img_resized)
    outputs=np.reshape(outputs, (21, 21, 256))
    
    p=np.zeros((21,21))
    somma=np.zeros((21,21))
    for i in range(w.shape[0]):
        p=w[i]*outputs[:, :, i]
        somma=somma+p

    print(somma.shape)

    plt.imshow(somma, cmap='seismic')
    plt.show()
    
    
```
